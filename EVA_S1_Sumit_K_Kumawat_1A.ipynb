```
# import the necessary packages
import numpy as np
from urllib.request import urlopen
import cv2
from matplotlib import pyplot as plt

# METHOD #1: OpenCV, NumPy, and urllib
def url_to_image(url):
	# download the image, convert it to a NumPy array, and then read
	# it into OpenCV format
	resp = urlopen(url)
	image = np.asarray(bytearray(resp.read()), dtype="uint8")
	image = cv2.imdecode(image, cv2.IMREAD_COLOR)
 
	# return the image
	return image
```

```
image = url_to_image("https://fontsarena-cd5e.kxcdn.com/wp-content/uploads/2019/04/helvetica-now-font-400x364.png")
```

#1. Horizontal Edge Detector with 3x3 Kernal

$$\begin{bmatrix}
1 & 1 & 0 \\\ -1 & -1 & 0 \\\ -1 & 0 & -1
\end{bmatrix}$$

```
#Horizotal
kernel = np.float32([[1,1,0],[-1,-1,0],[-1,0,-1]])


dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

#2. 45 Degree Angle Detector (either) with 3x3 Kernal

$$\begin{bmatrix}
1 & -1 & 0 \\\ 1 & -1 & 0 \\\ -1 & 0 & -1
\end{bmatrix}$$

```
#45 Degree Angle lines
kernel = np.float32([[1,-1,0],[1,-1,0],[-1,0,-1]])


dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

#3. Blur Kernel with 3x3 Kernal

$$\begin{bmatrix}
0 & -1 & 0 \\\ 1 & 1 & 1 \\\ 0 & -1 & 0
\end{bmatrix}$$

```
#Blur
kernel = np.float32([[0,-1,0],[1,1,1],[0,-1,0]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

#4. Sharpen Kernel with 3x3 Kernal

$$\begin{bmatrix}
-1 & -1 & 0 \\\ 1 & 1 & 0 \\\ 1 & 1 & -1
\end{bmatrix}$$

```
#Sharpen
kernel = np.float32([[-1,-1,0],[1,1,0],[1,1,-1]])


dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```
